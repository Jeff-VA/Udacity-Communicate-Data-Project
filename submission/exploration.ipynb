**Step 2:** The resulting violinplots in two facet grids are honestly overwhelmeng and are not as insightful as initially thought.  It is hard to tell whether the mode values around 35% change depending on any other variables.

In turn, side by side heat maps of grouped APR modes and means for each two compared qualitative variables would be useful.  Since it would be hard to code each of these heat map visualizations separately, a single function is created and used four times below.

```python
def modemean_heatmap(cat_var_l, cat_var_2):
    '''
    Takes two categorical variable inputs that are column variables of df_trimmed.  
    The BorrowerAPR collumn values are then grouped by the two inputed categorical
    variables by mode and by mean.  The mean and mode variables are then applyed to 
    two heatmap multivariate charts.
    
    The resulting heatmaps are returned as a single subplotted figure with the 
    'Mode APR' heatmap on the left and 'Mean APR' heatmap on the right.
    '''
    #assigns mode APR of grouped variables to cat_modes
    mode = lambda x:x.value_counts().index[0]
    cat_modes = df_trimmed.groupby([cat_var_l, cat_var_2])['BorrowerAPR'].agg(mode)
    cat_modes = cat_modes.reset_index(name = 'mode')
    cat_modes = cat_modes.pivot(index = cat_var_l, columns = cat_var_2, values = 'mode')
    
    #assigns mean APR of grouped variables to cat_means
    cat_means = df_trimmed.groupby([cat_var_l, cat_var_2]).mean()['BorrowerAPR']
    cat_means = cat_means.reset_index(name = 'mean')
    cat_means = cat_means.pivot(index = cat_var_l, columns = cat_var_2, values = 'mean')
    
    #creates two by one subplotted heatmaps of cat_means and cat_modes
    fig, ax = plt.subplots(ncols=2, figsize=(15,8))
    ax1 = sns.heatmap(cat_modes, annot = True, fmt = '.3f',
                      cmap = 'Blues', vmin=0.1, vmax=0.38, ax = ax[0])
    ax2 = sns.heatmap(cat_means, annot = True, fmt = '.3f',
                      cbar_kws = {'label' : 'APR'},
                      cmap = 'Blues', vmin=0.1, vmax=0.38, ax = ax[1])
    ax1.title.set_text('Mode APR')
    ax2.title.set_text('Mean APR')
    plt.ylabel('')
    return fig
```

```python
#plots heatmaps of mode and mean `BorrowerAPR by `EmploymentStatus` and `Term`
modemean_heatmap('EmploymentStatus', 'Term');
```
